```python
df["category"] = df["category"].map(category_mapping)
```

```python
df["category"] = df["category"].fillna("General")
```

```python
df["category"].value_counts()
```

```python
ax = df["category"].value_counts().plot(kind="bar" , legend=False)
ax.set_title("Ticket Category Distribution")
ax.set_xlabel("")
plt.show()
```

```python
df["text_length"] = df["ticket_text"].apply(len)
```

```python
sns.histplot(df["text_length"], bins=50)
plt.title("Ticket Text Length Distribution")
plt.show()
```

```python
sns.histplot(df["text_length"], bins=50)
plt.title("Ticket Text Length Distribution")
plt.axvline(700, color='red', linestyle='--', label='Tail threshold')
plt.xlabel("")
plt.legend()
plt.show()
```
